```python
#pairplots
#based on category
sns.pairplot(df, hue='Category')
plt.show()
```

```python
df.hist(figsize=(15,10), bins=50)
plt.show()
```

```python
#check for correlation
df.corr()
```

```python
#heatmap
corr=df.corr()
sns.heatmap(corr,annot=True)
```

```python
#countplots
sns.countplot(df['Category'])
plt.xticks(rotation=90)
plt.title('Category',fontsize=20)
```

```python
sns.countplot(df['Sub-Category'])
plt.xticks(rotation=90)
plt.title('Sub-Category',fontsize=10)
```

```python
fig, axs=plt.subplots(ncols=2)
sns.countplot(df['Ship Mode'], ax=axs[0])
sns.countplot(df['Segment'], ax=axs[1])
axs[0].set_title('Ship Mode',fontsize=10)
axs[1].set_title('Segment',fontsize=10)
plt.xticks(rotation=50)
plt.show()
```

```python
grouped=df.groupby(['Category','Sub-Category'])
grouped.size()
```

```python
#category vs sub-category
plt.subplots(figsize=(20,15))
sns.countplot(x='Category',hue='Sub-Category',data=df)
plt.title('Category Vs. Sub-Category',fontsize=10)
```

```python
df['Country'].value_counts()
```

```python
#state analysis
df1=df['State'].value_counts()
df1
```

```python
df['State'].value_counts().plot(kind='bar',figsize=(20,15))
plt.ylabel('Frequency/Number of deals')
plt.xlabel('States')
plt.title('State Wise Dealings',fontsize=10)
plt.show()
```

```python
df['State'].value_counts().mean()
```

```python
df['Segment'].value_counts()
```

```python
#segment wise analysis of sales,discount and profit
df_segment=df.groupby(['Segment'])[['Sales','Discount','Profit']].mean()
df_segment
```

```python
plot=df_segment.plot.pie(subplots=True,figsize=(18,20),autopct='%1.1f%%' ,labels=df_segment.index)
```

```python
#state wise analysis of sales,discount and profit
df_state=df.groupby(['State'])[['Sales','Discount','Profit']].mean()
df_state
```

```python
plt.figure(figsize=(20,20))
plt.pie(df_state['Sales'],labels=df_state.index)
plt.title('State Wise Sales Analysis',fontsize=10)
plt.legend()
plt.xticks(rotation=90)
plt.show()
```

```python
df_state.sort_values('Profit')[['Profit']].plot(kind='bar',figsize=(15,10))
plt.title('State wise Profit analysis',fontsize=10)
plt.ylabel('Profit per State')
plt.xlabel('States')
plt.show()
```

```python
df_state.sort_values('Discount')[['Discount']].plot(kind='bar',figsize=(15,10))
plt.title('State wise Discount analysis',fontsize=10)
plt.ylabel('Discount per state')
plt.xlabel('States')
plt.show()
```

```python
#region wise sales,discount and profit analysis
df_region=df.groupby(['Region'])[['Sales','Discount','Profit']].mean()
df_region
```

```python
plot=df_region.plot.pie(subplots=True,figsize=(18,20),labels=df_region.index)
```

```python
#category wise sales,discount and profit analysis
df_Category=df.groupby(['Category'])[['Sales','Discount','Profit']].mean()
df_Category
```

```python
plot=df_Category.plot.pie(subplots=True,figsize=(18,20),labels=df_Category.index)
```

```python
#sub-category wise sales,discounts and profit analysis
df_subcategory=df.groupby(['Sub-Category'])[['Sales','Discount','Profit']].mean()
df_subcategory
```

```python
plt.figure(figsize=(20,25))
plt.pie(df_subcategory['Sales'],labels=df_subcategory.index)
plt.title('Sub-Category Wise Sales Analysis',fontsize=10)
plt.legend()
plt.xticks(rotation=90)
plt.show()
```
